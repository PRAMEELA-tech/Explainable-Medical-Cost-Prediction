```python
# PHASE M5 â€” Feature Engineering (Advanced but Safe)

import pandas as pd
import numpy as np

print("Phase M5 started successfully")
```

```python
# Load cleaned dataset from Phase M4
df = pd.read_csv("../data/insurance.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["bmi_category"] = df["bmi"].apply(bmi_category)

df["bmi_category"].value_counts()
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["age"].apply(age_group)

df["age_group"].value_counts()
```

```python
df["smoker_binary"] = df["smoker"].map({"yes": 1, "no": 0})

df[["smoker", "smoker_binary"]].head()
```

```python
df["age_smoker_interaction"] = df["age"] * df["smoker_binary"]

df[["age", "smoker", "age_smoker_interaction"]].head()
```

```python
df["bmi_smoker_interaction"] = df["bmi"] * df["smoker_binary"]

df[["bmi", "smoker", "bmi_smoker_interaction"]].head()
```

```python
df.drop(columns=["smoker_binary"], inplace=True)
```

```python
df.to_csv("../data/insurance_feature_engineered.csv", index=False)

print("Feature-engineered dataset saved successfully")
```
